```python
# %load "test_myOLS.py"
from __future__ import print_function
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.sandbox.regression.predstd import wls_prediction_std

import pandas as pd
filenm = 'nerlove1963.csv'

hsb2 = pd.read_csv(filenm)

print(hsb2.head(10))
print("====print tail()======")
print(hsb2.tail())
hsb2.info()

## print(hsb2)
#lntc=y, x=lnq,lnpl,lnpk,lnpf

X=np.column_stack((hsb2['lnq'],hsb2['lnpl'],hsb2['lnpk'],hsb2['lnpf']))
X=sm.add_constant(X)
model=sm.OLS(hsb2['lntc'],X)
results=model.fit()
print(results.summary())

print('Parameters: ', results.params)
print('R2: ', results.rsquared)
```
